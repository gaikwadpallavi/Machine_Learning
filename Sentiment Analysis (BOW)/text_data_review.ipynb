```python
sprint.sort_values(['Time'], ascending=True, inplace=True)
sprint.head()
```

#### Dropping the unwanted columns from our data frame

```python
sprint.drop("ProductId", inplace=True, axis=1)
sprint.drop("ProfileName", inplace=True, axis=1)
sprint.drop("HelpfulnessNumerator", inplace=True, axis=1)
sprint.drop("HelpfulnessDenominator", inplace=True, axis=1)
sprint.drop("Time", inplace=True, axis=1)


```

```python
sprint.head()
```

 #### Score' greater than 3 equal to +ve to -ve class


```python
score = []
for i in sprint['Score']:
    if i < 3:
        score.append('0')
    else:
        score.append('1')  
sprint['Score'] = score  
```

#### model building

```python
 from sklearn.model_selection import train_test_split
```

```python
total_size=len(sprint)

train_size=int(0.70*total_size)

#training dataset
train=sprint.head(train_size)
#test dataset
test=sprint.tail(total_size - train_size)
```

####  Removing all rows where 'Score' is equal

```python
train = train[train.Score != 3]
test = test[test.Score != 3]
```

### text preprocessing

```python
lst_text = train['ReviewText'].tolist()

lst_summary = train['ReviewSummary'].tolist()
```

```python
lst_text
```

```python
lst_summary
```

```python
test_text = test['ReviewText'].tolist()
test_text
```

### coverting lower case

```python
lst_text = [str(item).lower() for item in lst_text]
lst_summary = [str(item).lower() for item in lst_summary]
```

```python
test_text = [str(item).lower() for item in test_text]
```

#### Removing the html tags

```python
import re
def striphtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('', data)

for i in range(len(lst_text)):
    lst_text[i] = striphtml(lst_text[i])
    lst_summary[i] = striphtml(lst_summary[i])
```

```python
for i in range(len(test_text)):
    test_text[i] = striphtml(test_text[i])
```

### stemming 

```python
%time
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
for i in range(len(lst_text)):
    text_filtered = []
    summary_filtered = []
    text_word_tokens = []
    summary_word_tokens = []
    text_word_tokens = lst_text[i].split()
    summary_word_tokens = lst_summary[i].split()
    for r in text_word_tokens:
        text_filtered.append(str(stemmer.stem(r)))
    lst_text[i] = ' '.join(text_filtered)
    for r in summary_word_tokens:
        summary_filtered.append(str(stemmer.stem(r)))
    lst_summary[i] = ' '.join(summary_filtered)
```

```python
for i in range(len(test_text)):
    text_filtered = []
    text_word_tokens = []
    text_word_tokens = test_text[i].split()
    for r in text_word_tokens:
        if not r in stop_words:
            text_filtered.append(str(stemmer.stem(r)))
    test_text[i] = ' '.join(text_filtered)
```

#### Converting text to numerical vectors - BOW representation

```python
from sklearn.feature_extraction.text import CountVectorizer
vocab = CountVectorizer()
train_bow = vocab.fit_transform(lst_text)
 
```

```python
train_bow
```

```python
X_test_dtm = vocab.transform(test_text)
X_test_dtm
```

### multinominal navie bayes

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train_bow, train.Score)
```

```python
y_pred_class_nb = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
metrics.accuracy_score(test.Score, y_pred_class_nb)
```

```python
con_metrics_nb = metrics.confusion_matrix(test.Score, y_pred_class_nb)
con_metrics_nb
```
